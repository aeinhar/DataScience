### loc

```python
df.loc[0:4, ['Name', 'Age', 'Ticket']]
```

```python
# select all rows with a condition
df.loc[df.Age >= 25]
```

```python
# select with multiple conditions
df.loc[(df.Age <= 40) & (df.Embarked == 'S') & (df.Sex=='female')]
```

```python
#slice
df.loc[6:30]
```

```python
# select few columns with a condition
df.loc[(df.SibSp == 1), ['Ticket', 'Fare']]
```

```python
# update a column with condition
df.loc[(df.Age >= 35), ['Sex']] = 'female'
df
```

```python
# update multiple columns with condition
df.loc[(df.Pclass == 3), ['Sex', 'Parch']] = ['male','0']
df
```

### iloc

```python
# select rows with indexes
df.iloc[[0,4]]
```

```python
# select rows with particular indexes and particular columns
df.iloc[[3,7],[1,3]]
```

```python
# select a range of rows
df.iloc[1:5]
```

```python
# select a range of rows and columns
df.iloc[1:4,2:7]
```
